```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as ply
```

```python
# Importar el separador de muestras para entrenamiento y pruebas
from sklearn .model_selection import train_test_split
```

```python
# Cargar el dataset
data = sns.load_dataset('iris')
data.head(5)
```

```python
# Resumen de datos
data.describe()
```

```python
# Distribución de cada clase
data.groupby('species').size()
```

```python
# Separación de datos en Entrenamiento y Prueba
train, test = train_test_split(data, test_size=0.4, stratify=data['species'], random_state=42)
```

```python
train.head()
```

```python
## Análisis exploratorio de datos
```

```python
# Histogramas para observar la distribución de 4 atributos
n_bins = 10
fig, axs = ply.subplots(2,2)
axs[0,0].hist(train['sepal_length'], bins=n_bins)
axs[0,0].set_title('Longitud de sépalo')
axs[0,1].hist(train['sepal_width'], bins=n_bins)
axs[0,1].set_title('Ancho de sépalo')
axs[1,0].hist(train['petal_length'], bins=n_bins)
axs[1,0].set_title('Longitud de petalo')
axs[1,1].hist(train['petal_width'], bins=n_bins)
axs[1,1].set_title('Ancho de petalo')
fig.tight_layout(pad=1.0);
```

```python
# Diagramas Box-plot
fig1, axsl = ply.subplots(2,2)
factores = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
categorias = ['setosa', 'versicolor', 'virginica']
sns.boxenplot(x='species', y='sepal_length', data=train, order=categorias, ax=axsl[0,0]) 
sns.boxenplot(x='species', y='sepal_width', data=train, order=categorias, ax=axsl[0,1]) 
sns.boxenplot(x='species', y='petal_length', data=train, order=categorias, ax=axsl[1,0]) 
sns.boxenplot(x='species', y='petal_width', data=train, order=categorias, ax=axsl[1,1]) 
fig1.tight_layout(pad=1.0);
```

```python
# Diagrama de violín
sns.violinplot(x='species', hue='species', y='petal_length', data=train, order=categorias)
```
