```python
import numpy as np 
import seaborn as sns
import statsmodels.api as sm
import pandas as pd 
from scipy import stats
```

```python

data = sns.load_dataset('iris')
data.head(10)
```

```python
data['species'].unique()
```

```python

data['species'].value_counts()
```

```python
x = data['petal_length']
y = data['petal_width'] 
```

```python

resultado = sm.OLS(y, x).fit()
```

```python
resultado.summary()
```

```python
resultado.params
```

```python
resultado.rsquared
```

```python

import math
```

```python
math.sqrt(resultado.rsquared)
```

```python
ax = sns.regplot(data=data, x='petal_length', y='petal_width')
ax.set_xlabel('Longitud de pétalo')
ax.set_ylabel('Ancho de pétalo')
ax.set_title('Relación entre la Longitud y el Ancho del pétalo')
ax.plot;
```
