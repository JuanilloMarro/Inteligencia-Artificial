```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.4)
x
```

```python
plt.plot(x, x, 'g--');
```

```python

y1 = x
y2 = x**2 # x^2
y3 = x**3 # x^3
```

```python
plt.plot(x, y1, 'g--', x, y2, 'bs', x, y3, 'r^')
plt.show()
```
